```python
from TTS.api import TTS
from TTS.tts.models.xtts import Xtts
from TTS.tts.configs.xtts_config import XttsConfig

import torch
import os
```

```python
def load_model(base_path="./models/"):
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

    model_path    = os.path.join(base_path, 'model.pth')
    config_path   = os.path.join(base_path,'config.json')
    vocab_path    = os.path.join(base_path,'vocab.json')
    speaker_path  = os.path.join(base_path,'speakers_xtts.pth')

    config = XttsConfig()

    #load configuration
    config.load_json(config_path)

    #adjust parameters if needed

    #load the model
    tts = Xtts.init_from_config(config)
    tts.load_checkpoint(config, 
                        checkpoint_path=model_path, 
                        vocab_path=vocab_path, 
                        speaker_file_path=speaker_path,
                        eval=True)
    
    tts = tts.to(device)

    return tts
```

```python
tts = load_model()
```

```python
#base_path = "C:/Users/Ruud/AppData/Local/tts/tts_models--multilingual--multi-dataset--xtts_v2/"

base_path = "./models/"


model_path  = os.path.join(base_path, 'model.pth')
config_path = os.path.join(base_path,'config.json')
vocab_path = os.path.join(base_path,'vocab.json')
voice_path = os.path.join(base_path,'speakers_xtts.pth')

config = XttsConfig()
#config.models_dir = os.path.join(models_dir,'tts')
config.load_json(config_path)
```

```python
config['temperature'] = 1.0
config
```

```python
tts = Xtts.init_from_config(config)
tts.load_checkpoint(config, checkpoint_path=model_path, vocab_path=vocab_path, speaker_file_path=voice_path,eval=True)
```

```python
selected_voice = 'Bert'

audio_path = './references/'

audio_path = os.path.join(audio_path, selected_voice + '.wav')


gpt_cond_latent, speaker_embedding = tts.get_conditioning_latents(audio_path=audio_path)
```

```python
tts = tts.to(device)
```

```python
output = tts.inference("hello, how are you doing today? I'm doing swimmingly!", 
                     gpt_cond_latent=gpt_cond_latent,
                     speaker_embedding=speaker_embedding,
                     language='en')
```
